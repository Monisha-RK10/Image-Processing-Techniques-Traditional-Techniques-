```python
#detect objects in the image via HSV (mask and filter) by color trackbars
import cv2
import numpy as np
image_path = r'D:\ComputerVision_Projects\Image_Processing_1\input\color_balls.jpg'
image = cv2.imread(image_path)
image = cv2.resize(image,(300,300))

#create an empty function 
def nothing(x):
    pass

#create options and a window to store those options
cv2.namedWindow('object detection')
cv2.createTrackbar('switch','object detection', 0,1, nothing)
cv2.createTrackbar('h_lower','object detection', 0,255, nothing)
cv2.createTrackbar('s_lower','object detection', 0,255, nothing)
cv2.createTrackbar('v_lower','object detection', 0,255, nothing)
cv2.createTrackbar('h_upper','object detection', 255,255, nothing)
cv2.createTrackbar('s_upper','object detection', 255,255, nothing)
cv2.createTrackbar('v_upper','object detection', 255,255, nothing)

while True:
    #convert the image from bgr to hsv
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    #access the trackbar position for the switch
    switch = cv2.getTrackbarPos('switch','object detection')
    
    if cv2.waitKey(1) == ord('q'):
        break
        
    #implement the code as per the switch's status
    if switch == 0:
        cv2.imshow('image', image)
    else:
        
        #access the status of the lower and upper ranges of hsv
        h_lower = cv2.getTrackbarPos('h_lower','object detection')
        s_lower = cv2.getTrackbarPos('s_lower','object detection')
        v_lower = cv2.getTrackbarPos('v_lower','object detection')
        h_upper = cv2.getTrackbarPos('h_upper','object detection')
        s_upper = cv2.getTrackbarPos('s_upper','object detection')
        v_upper = cv2.getTrackbarPos('v_upper','object detection')
    
        #set the lower and upper limits for the mask and pass them to the 'inRange' function
        lower_hsv = np.array([h_lower, s_lower, v_lower])
        upper_hsv = np.array([h_upper, s_upper, v_upper]) 
        image_mask = cv2.inRange(image_hsv, lower_hsv, upper_hsv)
        
        #filter the image with the mask
        image_filter = cv2.bitwise_and(image,image,mask=image_mask)
        cv2.imshow('image_filter', image_filter)
        cv2.imshow('image_mask', image_mask)
    
cv2.destroyAllWindows()
```

```python
#detect objects in a video using HSV and color trackbars
import cv2
import numpy as np

#capture video from the camera
video = cv2.VideoCapture(0)

#create a bin to store the output
fourcc = cv2.VideoWriter_fourcc(*"XVID")
output = cv2.VideoWriter("D:\ComputerVision_Projects\Image_Processing_1\output\object_detection.avi", fourcc, 20.0, (400,400))

#create an empty function
def nothing(x):
    pass

#create a window to hold the options for the HSV
cv2.namedWindow('ObjectDetection')

#create trackbars for the HSV
cv2.createTrackbar('switch','ObjectDetection',0,1,nothing)
cv2.createTrackbar('lower_h','ObjectDetection',0,255,nothing)
cv2.createTrackbar('lower_s','ObjectDetection',0,255,nothing)
cv2.createTrackbar('lower_v','ObjectDetection',0,255,nothing)
cv2.createTrackbar('upper_h','ObjectDetection',255,255,nothing)
cv2.createTrackbar('upper_s','ObjectDetection',255,255,nothing)
cv2.createTrackbar('upper_v','ObjectDetection',255,255,nothing)

while video.read():
    ret, frame = video.read()
    if ret == True:
        frame = cv2.resize(frame,(400,400))
        frame_hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        output.write(frame)
        
        #access the status of the switch
        switch = cv2.getTrackbarPos('switch','ObjectDetection')
        
        #implement the code for what happens when switch is ON/OFF
        if switch == 0:
            cv2.imshow("frame",frame)
        else:
            
            #get the status of all the options of HSV
            lower_h = cv2.getTrackbarPos('lower_h','ObjectDetection')
            lower_s = cv2.getTrackbarPos('lower_s','ObjectDetection')
            lower_v = cv2.getTrackbarPos('lower_v','ObjectDetection')
            upper_h = cv2.getTrackbarPos('upper_h','ObjectDetection')
            upper_s = cv2.getTrackbarPos('upper_s','ObjectDetection')
            upper_v = cv2.getTrackbarPos('upper_v','ObjectDetection')
            
            #create a mask
            lower_range = np.array([lower_h, lower_s, lower_v])
            upper_range = np.array([upper_h, upper_s, upper_v])
            frame_mask = cv2.inRange(frame_hsv, lower_range, upper_range)
            
            #filter the frame with the mask
            frame_filter = cv2.bitwise_and(frame, frame, mask=frame_mask)
            cv2.imshow("frame",frame)
            cv2.imshow("frame_filter",frame_filter)
            
        if cv2.waitKey(1) == ord('q'):
            break
            
#release output and window
video.release()
output.release()
cv2.destroyAllWindows()
```

```python
#blend two images via trackbar

#import libraries
import cv2
import numpy as np

#load abd resize images to same size
image_path1 = r'D:\ComputerVision_Projects\Image_Processing_1\input\avengers.jpg'
image_path2 = r'D:\ComputerVision_Projects\Image_Processing_1\input\color_balls.jpg'
image1 = cv2.imread(image_path1)
image2 = cv2.imread(image_path2)
image1 = cv2.resize(image1, (400,300))
image2 = cv2.resize(image2, (400,300))

#create an empty function
def nothing(x):
    pass

#create a window to store all the options
cv2.namedWindow('Blending with Tracker')
cv2.resizeWindow('Blending with Tracker', 400, 400)

#create all the options
cv2.createTrackbar('ON/OFF', 'Blending with Tracker', 0, 1, nothing)
cv2.createTrackbar('weight', 'Blending with Tracker', 1, 100, nothing)

while True:
    #access the status of all the options
    switch = cv2.getTrackbarPos('ON/OFF', 'Blending with Tracker')
    weight = cv2.getTrackbarPos('weight', 'Blending with Tracker')
    weight_percent = float(weight/100)
    
    if cv2.waitKey(1) == ord('q'):
        break
    
    #implement the code that you want to do with the options' status
    if switch == 1:
        image = cv2.addWeighted(image1, weight_percent, image2, 1-weight_percent, 0)
        image = cv2.putText(image, str(weight_percent), (50,50), cv2.FONT_ITALIC, 1, (2,240,5),4, cv2.LINE_AA)
    else:
        image = np.zeros((400,300,3), np.uint8)
        
    #plot the result based on the switch
    cv2.imshow('image', image)

#release all windows
cv2.destroyAllWindows()
```
